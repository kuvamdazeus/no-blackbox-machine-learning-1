```python
import csv
import matplotlib.pyplot as plt
```

```python
data = {
    "beds": [],
    "baths": [],
    "sizes": [],
    "prices": []
}

house_info = []

f = open('./houses/train.csv', 'r')
reader = csv.reader(f)

# skip the first row & put the reader cursor on the next row as it is the title of the csv file
reader.__next__()

for row in reader:
    size = row[2]
    price = row[-1]
    bed = row[0]
    bath = row[1]

    if (size == '' or price == '' or bed == '' or bath == ''):
        print("Invalid data")

    house_info.append([size, price])

    data["beds"].append(bed)
    data["baths"].append(bath)
    data["sizes"].append(size)
    data["prices"].append(price)

data.keys()
```

```python
len(house_info)
```

```python
plt.scatter(data["sizes"], data["prices"])
plt.xlabel("House Size (Sq. ft)")
plt.ylabel("House Prices")

# turn off the x & y axis numbers
plt.xticks([])
plt.yticks([])
```
